```python
import nltk

import pandas as pd
# pandas is used for:
#     handling tables/data
#     CSV files
#     Excel files
#     datasets

import matplotlib.pyplot as plt
# Matplotlib: Graph drawing tool
```

```python
tokens = [
'Artificial', 'intelligence', 'is', 'transforming', 'industries',
'across', 'the', 'world', 'Students', 'are', 'learning',
'machine', 'learning', 'natural', 'language', 'processing',
'and', 'deep', 'learning', 'to', 'build', 'smart',
'applications', 'Businesses', 'use', 'AI', 'for',
'automation', 'customer', 'support', 'recommendation',
'systems', 'fraud', 'detection', 'and', 'data',
'analysis', 'Modern', 'AI', 'systems', 'can',
'understand', 'speech', 'recognize', 'images',
'generate', 'text', 'and', 'assist', 'humans',
'efficiently'
]
```

```python
unigrams = pd.Series(nltk.ngrams(tokens, 1)).value_counts()
# nltk.ngrams(tokens, 1): Creates unigrams.
# pd.Series(...): Converts data into a pandas Series.
    # 0       (I,)
    # 1    (love,)
    # 2      (AI,)
    # 3    (love,)

print(unigrams)
```

```python
unigrams[:10].sort_values().plot.barh(color="lightsalmon", width=.9, figsize=(12, 8))
# This line creates a horizontal bar graph of the top 10 unigram frequencies using pandas and Matplotlib.
```

```python
bigrams = pd.Series(nltk.ngrams(tokens, 2)).value_counts()
print(bigrams)
```

```python
trigrams = pd.Series(nltk.ngrams(tokens, 3)).value_counts()
print(trigrams)
```
