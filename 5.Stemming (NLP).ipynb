```python
# Stemming means:Reducing a word to its root/base form.
```

```python
from nltk.stem import PorterStemmer
# Imports the Porter Stemmer from NLTK.
# PorterStemmer is one of the most popular stemming algorithms in NLP.

ps = PorterStemmer()

connect_tokens = ['connecting', 'connected', 'connect', 'connects']

for t in connect_tokens:
    print(t,":", ps.stem(t))
```
