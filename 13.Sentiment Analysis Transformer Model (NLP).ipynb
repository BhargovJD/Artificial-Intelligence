```python
sentence = 'i had a great time at the movie it was really funny'
```

```python
import transformers
# Transformers is a popular Python library used for:
    # NLP (Natural Language Processing)
    # AI models
    # LLMs (Large Language Models)
    # text generation
    # translation
    # summarization
    # sentiment analysis

from transformers import pipeline
# This imports only the pipeline feature from the transformers library.
# pipeline() is a ready-made AI tool.
# It allows you to use powerful AI models with just a few lines of code.
```

```python
sentiment_pipeline = pipeline('sentiment-analysis')
# Loads pretrained sentiment AI model
```

```python
print(sentence)
```

```python
sentiment_pipeline(sentence)
```

```python
specific_model = pipeline(
    'sentiment-analysis',
    model='finiteautomata/bertweet-base-sentiment-analysis'
)

# 'sentiment-analysis' => “We want a sentiment analysis task.”
# model => “Do not use the default model. Use this specific model instead.”
```

```python
specific_model(sentence)
```
