```python
from sklearn.feature_extraction.text import TfidfVectorizer
# 1. sklearn
    # sklearn = Scikit-learn
    # A popular Python library used for:
        # Machine Learning
        # AI
        # NLP

# 2. feature_extraction
    # AI cannot understand raw text directly.
    # So we extract useful information called features.
    # Example features:
        # words
        # word counts
        # important words

# 3. text
    # Means we are working with text data.

# 4. TfidfVectorizer
    # This is the main tool.
    # It converts text into numbers using:
        # Bag of Words
        # TF-IDF importance scoring

# It performs TWO tasks together.
    # 1	Count words
    # 2	Calculate word importance

import pandas as pd

# Sample sentences
sentences = [
    "AI is amazing",
    "AI is powerful",
    "AI is future"
]

# Create TF-IDF object
tfidf = TfidfVectorizer()

# Learn vocabulary + calculate TF-IDF values
tfidf_fit = tfidf.fit_transform(sentences)

# Step 1 → fit()
    # The model first learns vocabulary.
    # It scans all sentences and finds unique words.
# Step 2 → transform()
    # Now it converts sentences into TF-IDF values.
    # "AI is amazing"
    # [0.38 0.72 0.00 0.38 0.00]

# Convert into readable table
tfidf_table = pd.DataFrame(
    tfidf_fit.toarray(),
    columns=tfidf.get_feature_names_out()
)

# Print result
print(tfidf_table)



# Common words:
    # ai
    # is
        # get lower scores.

# Rare words:
    # amazing
    # powerful
    # future
        # get higher scores.
```
