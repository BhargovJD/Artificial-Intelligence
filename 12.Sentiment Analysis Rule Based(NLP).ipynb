```python
sentence = 'i had a great time at the movie it was really funny'
```

```python
#TextBlob Package
```

```python
from textblob import TextBlob
# Bring the TextBlob class from the textblob library into our program.
```

```python
sentiment_score = TextBlob(sentence)
# TextBlob(sentence) converts the normal text into a TextBlob object

print(sentiment_score.sentiment)
# sentiment_score =>	Variable
# .sentiment =>	Attribute/Property
# .polarity =>	Attribute/Property
# Extracts only the polarity value.
# Polarity tells whether the sentence is:
    # positive
    # negative
    # neutral

# Polarity =>	Positive vs Negative emotion
# Subjectivity =>	Opinion vs Fact

# Polarity	-1	+1	Negative → Positive
# Subjectivity	0	1	Fact → Opinion
```

```python
# VADER 
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
# Imports the SentimentIntensityAnalyzer class from the vaderSentiment library.
# VADER is a sentiment analysis tool specially designed for:
    # social media text
    # reviews
    # short sentences
    # emojis/slang

vader_sentiment = SentimentIntensityAnalyzer()
# Creates an object of the analyzer.

print(sentence)
print(vader_sentiment.polarity_scores(sentence))
```

```python
# TextBlob =>	Basic emotion checker
# VADER =>	Smart social-media emotion detector


# Use TextBlob when:
    # learning NLP
    # doing simple projects
    # analyzing clean/formal text

# Use VADER when:
    # analyzing reviews
    # social media posts
    # YouTube comments
    # tweets/chats
    # emotional short text
```
