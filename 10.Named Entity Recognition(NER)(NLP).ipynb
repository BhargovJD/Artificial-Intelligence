```python
import spacy
# spaCy is a powerful NLP (Natural Language Processing) library used for: POS tagging

from spacy import displacy
# Used to display objects/output properly in notebook.

from spacy import tokenizer
# Tokenizer splits text into smaller pieces called tokens.

from IPython.display import HTML, display
# Imports tools from Jupyter Notebook/IPython for displaying outputs beautifully.

import re
# Imports Python’s Regular Expression module.

import pandas as pd
    # Pandas is a Python library used for:
        # Data analysis
        # Data cleaning
        # Working with tables/spreadsheets
        # Handling CSV/Excel files
```

```python
nlp = spacy.load('en_core_web_sm')
# pretrained English NLP model in spaCy.
    # What This Model Can Do
        # After loading, the model can perform:
            # Tokenization
            # POS tagging
            # Named Entity Recognition (NER)
            # Lemmatization
            # Dependency parsing
```

```python
google_text = "Google was founded in 1998 by American computer scientists Larry Page and Sergey Brin. Together, they own about 14% of its publicly listed shares and control 56% of its stockholder voting power through super-voting stock. The company went public via an initial public offering (IPO) in 2004. In 2015, Google was reorganized as a wholly owned subsidiary of Alphabet Inc. Google is Alphabet's largest subsidiary and is a holding company for Alphabet's internet properties and interests. Sundar Pichai was appointed CEO of Google in 2015, replacing Larry Page, who became the CEO of Alphabet. In 2019, Pichai also became the CEO of Alphabet."
```

```python
print(google_text)
```

```python
spacy_doc = nlp(google_text)
print(spacy_doc)
```

```python
for word in spacy_doc.ents:
    print(word.text, word.label_)
    
# spacy_doc.ents
    # Loops through each entity one by one.
# word.text
    # Returns actual entity text.
# word.label_
    # Returns entity category/type.
    
# This code extracts and prints Named Entities from a sentence using spaCy.
# Named Entities are real-world names like:
    # Person names
    # Places
    # Organizations
    # Dates
    # Money
    # Countries
```

```python
html = displacy.render(spacy_doc,style="ent", jupyter=False)
display(HTML(html))

# 1. displacy.render()
    # displacy is spaCy’s visualization tool.
# 2. style="ent"
    # ent means:
    # Visualize Named Entities.
# 3. jupyter=False
    # Tells spaCy:
    # “Do not directly display in Jupyter notebook.”
    # Instead:
    # Generate HTML code only
    # Store it in variable
```
