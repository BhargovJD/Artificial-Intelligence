```python
customer_reviews = [
    "bhargo is a good boy",
    "bhargov come here",
    "ok ok :)",
    "needed one",
    "wanted one"
]
bhargov_reviews = []

pattern_to_find = r"bhargov?"
# The previous character is optional
# So this pattern means:
# "bhargo"
# "bhargov"

for string in customer_reviews:
    if(re.search(pattern_to_find,string)):
        bhargov_reviews.append(string)
print(bhargov_reviews)
```

```python
o_reviews = []
pattern_to_find = r"^o"
# Match strings that START with the letter o

for string in customer_reviews:
    if(re.search(pattern_to_find,string)):
        o_reviews.append(string)
print(o_reviews)
```

```python
y_reviews = []
pattern_to_find = r"y$"
# Match strings that END with the letter y

for string in customer_reviews:
    if(re.search(pattern_to_find,string)):
        y_reviews.append(string)
print(y_reviews)
```

```python
need_want_reviews = []
pattern_to_find = r"(need|want)ed"
# Match strings that have needed and wanted

for string in customer_reviews:
    if(re.search(pattern_to_find,string)):
        need_want_reviews.append(string)
print(need_want_reviews)
```

```python
no_punc_reviews = []
pattern_to_find = r"[^\w\s]"
# Find everything that is NOT a word character and NOT a space.

for string in customer_reviews:
    no_punc_string = re.sub(pattern_to_find,"",string)
    no_punc_reviews.append(no_punc_string)
print(no_punc_reviews)
```
