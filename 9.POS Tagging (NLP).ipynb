```python
# POS Tagging in AI/NLP stands for Part-of-Speech Tagging.
# It means identifying the grammar role of each word in a sentence.
    # “Book a flight.”: Here Book is a verb.
    # “This is a good book.”: Now book is a noun.
# POS tagging helps AI understand the correct meaning.
```

```python
import spacy
# spaCy is a powerful NLP (Natural Language Processing) library used for: POS tagging

from spacy import displacy
# Used to display objects/output properly in notebook.

from spacy import tokenizer
# Tokenizer splits text into smaller pieces called tokens.

from IPython.display import HTML, display
# Imports tools from Jupyter Notebook/IPython for displaying outputs beautifully.

import re
# Imports Python’s Regular Expression module.

import pandas as pd
    # Pandas is a Python library used for:
        # Data analysis
        # Data cleaning
        # Working with tables/spreadsheets
        # Handling CSV/Excel files
```

```python
nlp = spacy.load('en_core_web_sm')
# pretrained English NLP model in spaCy.
    # What This Model Can Do
        # After loading, the model can perform:
            # Tokenization
            # POS tagging
            # Named Entity Recognition (NER)
            # Lemmatization
            # Dependency parsing
```

```python
emma_js = 'emma Woodhouse handsome clever and rich with a comfortable home and happy disposition seemed to unite some of the best blessings of existence and had lived nearly twenty one years in the world with very little to distress or vex her she was the youngest of the two daughters of a most affectionate indulgent father and had in consequence of her sister’s marriage been mistress of his house from a very early period her mother had died too long ago for her to have more than an indistinct remembrance of her caresses and her place had been supplied by an excellent woman as governess who had fallen little short of a mother in affection'
```

```python
print(emma_js)
```

```python
spacy_doc = nlp(emma_js)
```

```python
pos_df = pd.DataFrame(columns=['token','pos_tag'])
# This line creates an empty Pandas DataFrame (table) with two column names.
```

```python
for token in spacy_doc:
    pos_df = pd.concat([
    pos_df,
    pd.DataFrame.from_records([
        {'token': token.text, 'pos_tag': token.pos_}
    ])
], ignore_index=True)

# This line adds a new row into the DataFrame pos_df.


# or faster version
rows = []

for token in spacy_doc:
    rows.append({
        'token': token.text,
        'pos_tag': token.pos_
    })

pos_df = pd.DataFrame(rows)
```

```python
print(pos_df)
```

```python
print(pos_df.head(10))
```

```python
pos_df_counts = pos_df.groupby(['token','pos_tag']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
# 1. groupby(['token','pos_tag'])
# 2. .size()
    # Counts rows in each group.
# 3. .reset_index(name='counts')
    # Converts grouped result into normal DataFrame.
    # Also renames count column to "counts".
# 4. .sort_values(by='counts', ascending=False)

print(pos_df_counts.head(10))
```

```python
pos_df_poscounts = pos_df_counts.groupby(['pos_tag'])['token'].count().sort_values(ascending=False)

# 1. pos_df_counts
    # This is your DataFrame.
# 2. .groupby(['pos_tag'])
    # Groups rows based on POS tags.
# 3. ['token']
    # Selects only the token column.
# 4. .count()
    # Counts how many tokens exist in each POS group.
# 5. .sort_values(ascending=False)
    # Sorts counts from highest to lowest.
```
