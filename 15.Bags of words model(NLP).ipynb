```python
import pandas as pd
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Importing the CountVectorizer tool from the scikit-learn library to convert text into numbers using the Bag of Words technique.
# 1. sklearn
    # sklearn = Scikit-learn
    # A popular Python library used for:
        # Machine Learning
        # AI models
        # NLP tasks
# 2. feature_extraction
    # AI cannot understand raw text.
    # So we extract features (important information) from text.
    # Example:
        # Sentence:
        # "I love AI"
        # Features can be:
        # love
        # AI
        # word counts
# 3. text
    # Means we are working with text data.
# 4. CountVectorizer
    # This is the main tool.
    # It converts sentences into a table of word counts.
    # It uses the Bag of Words model.
```

```python
data = [
    "I love AI",
    "AI is amazing"
]

count_vec = CountVectorizer()

count_vec_fit = count_vec.fit_transform(data)

print(count_vec_fit.toarray())

bag_of_words = pd.DataFrame(
    count_vec_fit.toarray(),
    columns = count_vec.get_feature_names_out()
)
# This line converts the Bag of Words output into a readable table (DataFrame).


print(bag_of_words)
```

```python


# Why “Bag” of Words?

# Because the sentence is treated like a bag full of words.

# The order does not matter.

# For AI:

# “I love AI”
# “AI love I”

# look almost the same.

# Real-Life Use in AI

# Bag of Words is used in:

# Spam detection
# Sentiment analysis
# Text classification
# Chatbots
# Review analysis

# Example:

# If words like:

# “great”
# “amazing”
# “awesome”

# appear many times → AI may predict positive review.

# Advantages

# ✅ Very simple
# ✅ Fast
# ✅ Easy for beginners

# Disadvantages

# ❌ Ignores meaning
# ❌ Ignores grammar
# ❌ Ignores word order
# ❌ Cannot understand context

# Example:

# “I am happy”
# “I am not happy”

# Bag of Words may think both are similar because both contain “happy”.
```
