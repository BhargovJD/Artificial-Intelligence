```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
```

```python
data = pd.read_csv("tripadvisor_hotel_reviews.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data["Review"][0]
```

```python
data["review_lowercase"] = data["Review"].str.lower()
```

```python
data.head()
```

```python
en_stopwords = stopwords.words('english')
en_stopwords.remove("not")
data['review_no_stopwords'] = data['review_lowercase'].apply(
    lambda x: ' '.join([word for word in x.split() if word not in en_stopwords])
)
# Step 1 — data['review_lowercase']
# Step 2 — .apply(): Runs a function on every row.
# Step 3 — lambda x: here, x = one review sentence at a time
# Step 4 — x.split(): Splits sentence into words. "i love AI".split() => ['i', 'love', 'AI']
# Step 5 — List Comprehension: [word for word in x.split() if word not in en_stopwords]
        # This loops through every word.
        # en_stopwords = ['is', 'a', 'the', 'this']
        # this is a good movie => good movie
        # i love AI => love AI
```

```python
data['review_no_stopwords'][0]
```

```python
data['review_no_stopwords_no_punct'] = data.apply(
    lambda x: re.sub(r"[*]", "star", x['review_no_stopwords']),
    axis=1
)

# For every review row, replace * symbols with the word star and save the cleaned sentence into a new column.
# 1. data.apply(): Runs a function row by row on the DataFrame.
# 2. lambda x: 
    # here, x = one row of the DataFrame
# 3. re.sub()
    # re.sub(pattern, replacement, text)
        # r"[*]"	find * symbol
        # "star"	replace with word "star"
        # text	sentence to process
# 4. axis=1
    # apply function row-wise
    # Without this:Pandas may process columns instead of rows.
    
```

```python
data.head()
```

```python
data['review_no_stopwords_no_punct'] = data.apply(
    lambda x: re.sub(r"[^\w\s]", "", x['review_no_stopwords']),
    axis=1
)

# For every review row, remove punctuation/special symbols from the text and save the cleaned sentence into a new column.
```

```python
data.head()
```

```python
data['tokenized'] = data.apply(
    lambda x: word_tokenize(x['review_no_stopwords_no_punct']),
    axis=1
)

# For every cleaned review sentence, split the sentence into individual word tokens and save them into a new column.
```

```python
data['tokenized']
data['tokenized'][0]
```

```python
ps = PorterStemmer()
data['stemmed'] = data['tokenized'].apply(lambda tokens:[ps.stem(token) for token in tokens])

# For every tokenized sentence, stem each word into its root form and save the stemmed words into a new column.

data.head()
```

```python
lemmatizer = WordNetLemmatizer()

data['lemmatized'] = data['tokenized'].apply(
    lambda tokens: [lemmatizer.lemmatize(token) for token in tokens]
)

# For every tokenized sentence, convert each word into its dictionary/root form and store the lemmatized words into a new column

data['lemmatized']
data['lemmatized'][0]
```

```python
tokens_clean = sum(data['lemmatized'],[])
# Take all token lists from every row and merge them into one big list of words.
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens_clean, 1)).value_counts())
# Create single-word groups from all cleaned tokens and count how many times each word appears.

print(unigrams)
```

```python
bigram = (pd.Series(nltk.ngrams(tokens_clean, 2)).value_counts())
print(bigram)
```

```python
trigram = (pd.Series(nltk.ngrams(tokens_clean, 3)).value_counts())
print(trigram)
```
