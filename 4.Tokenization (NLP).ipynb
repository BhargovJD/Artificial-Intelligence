```python
import nltk
# Imports the NLTK library. Natural Language Toolkit

nltk.download('punkt_tab')
# Downloads tokenizer data needed by NLTK.
# punkt is a pretrained tokenizer model used to split:
# sentences
# words

from nltk.tokenize import word_tokenize, sent_tokenize
# Imports two tokenizer functions.
```

```python
sentence = "Hello everyone. My name is bhargov"
sent_tokenize(sentence)
```

```python
sentence = "My name is bhargov"
word_tokenize(sentence)
```
