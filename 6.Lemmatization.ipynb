```python
# Lemmatization means: converting words into their actual dictionary/base form (called lemma).
# Unlike stemming, lemmatization tries to return a real meaningful word.
```

```python
import nltk
# Imports the NLTK library.

nltk.download('wordnet')
# WordNet is a large English dictionary/database used by NLTK.

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

like_tokens = ['likes', 'better', 'worse']
for t in like_tokens:
    print(t,":", lemmatizer(t))
```
