```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
# nltk → main library
# corpus → section containing datasets/text collections
# stopwords → one specific dataset inside corpus
```

```python
en_stopwords = stopwords.words('english')
print(en_stopwords)
```

```python
sentence = "I am drinking coffee in the morning"
sentence_no_stopwords = ' '.join([word for word in sentence.split() if word not in en_stopwords])
```

```python
# Take each word from the sentence and keep it only if it is NOT a stopword.
```

```python
print(sentence_no_stopwords)
```
